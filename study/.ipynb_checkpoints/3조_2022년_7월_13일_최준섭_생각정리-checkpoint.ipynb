```python
# 사용할 모듈 불러오기 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import re
```

```python
# 데이터 불러오기
items = pd.read_csv("./data/items.csv")
shops = pd.read_csv("./data/shops.csv")
train = pd.read_csv("./data/sales_train.csv")
test = pd.read_csv("./data/test.csv")
item_categories = pd.read_csv('./data/item_categories.csv')
```

```python
pivot_for_train = train.pivot_table(index = ['date_block_num'] , values=['item_cnt_day'],aggfunc='sum')
pivot_for_train
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(pivot_for_train.item_cnt_day)
```

```python
f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title('1st Order Differencing')
ax1.plot(pivot_for_train.item_cnt_day.diff())

ax2 = f.add_subplot(122)
plot_acf(pivot_for_train.item_cnt_day.diff().dropna(), ax=ax2)
plt.show()
```

```python
f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title('2nd Order Differencing')
ax1.plot(pivot_for_train.item_cnt_day.diff().diff())

ax2 = f.add_subplot(122)
plot_acf(pivot_for_train.item_cnt_day.diff().diff().dropna(), ax=ax2)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(pivot_for_train.item_cnt_day.dropna())
print('p-value: ', result[1])

result = adfuller(pivot_for_train.item_cnt_day.diff().dropna())
print('p-value: ', result[1])

result = adfuller(pivot_for_train.item_cnt_day.diff().diff().dropna())
print('p-value: ', result[1])
```
