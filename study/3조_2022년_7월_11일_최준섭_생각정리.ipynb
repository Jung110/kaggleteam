```python
def clean_text(inputString):
    text_rmv = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', inputString)
    return text_rmv
```

```python
shops['shop_name'] = shops["shop_name"].apply(clean_text)
shops
```

```python
shops.head()
```

```python
high_shops_name = shops.shop_name
high_shops_name.head()
```

```python
shops['shop_name'] = shops['shop_name'].str.lstrip()
shops.head()
```

```python
shops['city_name'] = shops['shop_name'].str.split(' ').str[0]
shops.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
shops['city_LE'] = le.fit_transform(shops['city_name'])
le.inverse_transform(shops['city_LE'] )
# 라벨인코딩
```
